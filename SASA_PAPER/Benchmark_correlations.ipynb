```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import numpy as np

# Read csv file and store in DataFrame
data = pd.read_csv('affinity_benchmark.csv')
print("Test")

plt.figure(figsize=(15,8))  # Set the size of the figure
x = data.iloc[:, 1].values[:, np.newaxis]  # Reshape X values into a 2D array

for i in range(data.iloc[:, 2:].values.shape[1]):
    y = data.iloc[:, i+2].values
    
    # Create linear regression model to fit the data
    model = LinearRegression()
    model.fit(x, y)  # Fit the X and y values to the model

    plt.scatter(range(len(y)), y, label=f"Column {i+1}")  # Plot scatter of data
    
    plt.plot(range(len(y)), model.predict(x), color='red', label="Fitted Line")  # Plot fitted line with same label

plt.legend()  # Add legend to the graph
plt.show()  # Display the figure
```
