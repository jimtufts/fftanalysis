```python
def opt(full_data):
    P = array([[2., 0., 0., 0., 0., 0.], 
               [0., 2., 0., 0., 0., 0.], 
               [0., 0., 2., 0., 0., 0.], 
               [0., 0., 0., 2., 0., 0.],
               [0., 0., 0., 0., 2., 0.], 
               [0., 0., 0., 0., 0., 2.]])
    q = array([-2., -2., -2., -2., -2., -2.]).reshape((6,))
    G = []
    h = []
    for i, data in enumerate(full_data):
        G.append(array([[data[0], data[8], 0., 0., 0., 0.], 
                        [data[1], 0., 0., data[9], 0., 0.], 
                        [0., data[2], data[10], 0., 0., 0.],
                        [0., 0., data[3], data[11], 0., 0.], 
                        [0., data[12], 0., 0., data[4], 0.], 
                        [data[5], 0., 0., 0., 0., data[13]], 
                        [0., 0., 0., data[14], data[6], 0.],
                        [0., 0., data[7], 0., 0., data[15]]]))
        h.append(array([2*data[16]-1, 2*data[17]-1, 2*data[18]-1, 2*data[19]-1, 
                        2*data[20]-1, 2*data[21]-1, 2*data[22]-1, 2*data[23]-1]).reshape((8,)))
    G = np.concatenate(G)
    h = np.concatenate(h)
    x = solve_qp(P, q, G, h)
    return x
```
