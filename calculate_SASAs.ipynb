```python
import mdtraj as md
import bpmfwfft.grids as grids
import numpy as np
# import pickle  ,.
```

```python
from os import walk
```

```python
f = {}
for (dirpath, dirnames, filenames) in walk('/home/jtufts/Desktop/FFT_PPI/2.redock/2.minimize'):
    if filenames[0] != 'run_me.sh':
        if 'complex.pdb' in filenames:
            f[dirpath.split('/')[7]] = dirpath + '/complex.pdb'
#             f.extend([dirpath + '/complex.pdb', dirpath.split('/')[8]])
```

```python
complex_sasas = {}
for key in f.keys():
    traj = md.load_pdb(f[key])
    print(f'doing sasa for {key}')
    complex_sasas[key] = md.shrake_rupley(traj)*100
```

```python
import pickle

pickle.dump(complex_sasas, open("/home/jtufts/src/p39/fftanalysis/data/complex_sasas_Angstroms.p", "wb"))
```

```python
complex_sasa_sums = {}
for key in complex_sasas.keys():
    complex_sasa_sums[key] = (np.sum(complex_sasas[key]))
    print(np.sum(complex_sasas[key]))
    
```

```python
rec_files = {}
for (dirpath, dirnames, filenames) in walk('/home/jtufts/Desktop/FFT_PPI/2.redock/1.amber'):
    if filenames[0] != 'run_me.sh':
        if 'receptor.prmtop' in filenames:
            rec_files[dirpath.split('/')[7]] = ['/home/jtufts/Desktop/FFT_PPI/2.redock/2.minimize/' + dirpath.split('/')[7] + '/receptor.inpcrd', dirpath + '/receptor.prmtop']
```

```python
print(rec_files)
```

```python
rec_sasas = {}
for key in rec_files.keys():
    print('from prmtop:' + rec_files[key][1])
    print('from inpcrd:' + rec_files[key][0])
    try:
        traj = md.load(rec_files[key][0], top=rec_files[key][1])
        print(f'doing sasa for {key} receptor')
        rec_sasas[key] = md.shrake_rupley(traj)*100
    except:
        print(f"SASA for {key} failed!")
        rec_sasas[key] = 0
```

```python
lig_files = {}
for (dirpath, dirnames, filenames) in walk('/home/jim/src/p27/FFT_PPI/2.redock/1.amber'):
    if filenames[0] != 'run_me.sh':
        if 'ligand.prmtop' in filenames:
            lig_files[dirpath.split('/')[8]] = ['/home/jim/src/p27/FFT_PPI/2.redock/2.minimize/' + dirpath.split('/')[8] + '/ligand.inpcrd', dirpath + '/ligand.prmtop']
```

```python
print(lig_files)
```

```python
lig_sasas = {}
for key in lig_files.keys():
    print('from prmtop:' + lig_files[key][1])
    print('from inpcrd:' + lig_files[key][0])
    try:
        traj = md.load(lig_files[key][0], top=lig_files[key][1])
        print(f'doing sasa for {key} ligand')
        lig_sasas[key] = md.shrake_rupley(traj)
    except:
        print(f"SASA for {key} failed!")
        lig_sasas[key] = 0
```

```python
pickle.dump(rec_sasas, open("/home/jim/src/p39/jupyter/fftanalysis/data/rec_sasas.p", "wb"))
pickle.dump(lig_sasas, open("/home/jim/src/p39/jupyter/fftanalysis/data/lig_sasas.p", "wb"))
```

```python
rec_sasa_sums = {}
for key in rec_sasas.keys():
    rec_sasa_sums[key] = (np.sum(rec_sasas[key]))
    print(f'{key} receptor sasa:', np.sum(rec_sasas[key]))
    
lig_sasa_sums = {}
for key in lig_sasas.keys():
    lig_sasa_sums[key] = (np.sum(lig_sasas[key]))
    print(f'{key} ligand sasa:', np.sum(lig_sasas[key]))
```

```python
pickle.dump(rec_sasa_sums, open("/home/jim/src/p39/jupyter/fftanalysis/data/rec_sasa_sums.p", "wb"))
pickle.dump(lig_sasa_sums, open("/home/jim/src/p39/jupyter/fftanalysis/data/lig_sasa_sums.p", "wb"))
```
