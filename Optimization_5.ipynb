```python
bad_systems = ["2HQS_A:H", "3A4S_A:D", "3K75_D:B", "3L5W_LH:I", "4IZ7_A:B"] 
def opt(full_data, cutoff):
    
    def get_index(label_list, label):
        i = np.where(label_list == label)
        return i[0][0]

    def get_pair_index(pair_list, pair):
        i = np.where(pair_list == pair)
        return i[0][0]

    def preprocess_data():
        atom_types = []
        atom_pairs = []
        system_names = []
        distances = []
        for row in full_data:
            if row[2] < cutoff and row[9] not in bad_systems:
                rl = row[5]
                ll = row[6]
                if row[5] == "CA" and row[7] != 12.01: #40.08
                    rl = "CA++"
                if row[6] == "CA" and row[8] != 12.01:
                    ll = "CA++"
                atom_types.extend([rl, ll])
                atom_pairs.append(np.array([rl,ll]))
                system_names.append(row[9])
                distances.append(row[2])
                
        return atom_types, atom_pairs, system_names, distances
        
    def count_all_set_occurrences(array):
        sorted_array = np.sort(array, axis=1)
        unique_rows = np.unique(sorted_array, axis=0)
        occurrences_dict = {}
    
        for target_set in unique_rows:
            count = np.count_nonzero(np.all(sorted_array == target_set, axis=1))
            occurrences_dict[tuple(target_set)] = count
        
        return occurrences_dict
    
    atom_types, atom_pairs, system_names, distances = preprocess_data()
    pair_dict = count_all_set_occurrences(np.array(atom_pairs))
    pair_dict_sorted = dict(sorted(pair_dict.items(), key=lambda item: item[1]))
    unique_types = np.unique(np.array(list(pair_dict_sorted.keys())))
    unique_types_count = unique_types.shape[0]
    type_pairs = np.array(list(pair_dict_sorted.keys()))
    type_pairs_count = type_pairs.shape[0]
    
    P = np.diag([2.]*type_pairs_count)
    q = np.full((type_pairs_count, 1), -2.)
    G = []
    h = []
    for d in data:
        if d[9] not in bad_systems and d[2] <= cutoff:
            G_1 = np.zeros((type_pairs_count, type_pairs_count))
            h_1 = np.zeros((type_pairs_count,))
            if d[5] == "CA" and d[7] == 40.08:
                d[5] = "CA++"
            if d[6] == "CA" and d[8] == 40.08:
                d[6] = "CA++"
            sigma1 = d[0]
            sigma1_ind = get_index(unique_types, d[5])
            sigma2 = d[1]
            try:
                sigma2_ind = get_index(unique_types, d[6])
            except:
                print(d[6])
            dist = d[2]
            pair_index = get_index(type_pairs, sorted([d[5], d[6]]))
            G_1[pair_index, sigma1_ind] = sigma1
            G_1[pair_index, sigma2_ind] = sigma2
            G.append(G_1)
            h_1[pair_index,] = 2*dist-1
            h.append(h_1)
    
    # print(list(z3.keys())[0],tuple(sorted((d[5], d[6]))),list(z3.keys())[0] == tuple(sorted((d[5], d[6]))))
    G = np.concatenate(G)
    h = np.concatenate(h)
    x = solve_qp(P, q, G, h, solver="quadprog")
    return x, pair_dict_sorted
```
