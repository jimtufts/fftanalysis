```python
import MDAnalysis as mda
```

```python
rec_prmtop = f"{ppi_path}/2.redock/1.amber/{system}/receptor.prmtop"
rec_inpcrd = f"{ppi_path}/2.redock/2.minimize/{system}/receptor.inpcrd"
lig_prmtop = f"{ppi_path}/2.redock/1.amber/{system}/ligand.prmtop"
lig_inpcrd = f"{ppi_path}/2.redock/2.minimize/{system}/ligand.inpcrd"
```

```python
receptor = mda.Universe(rec_prmtop, rec_inpcrd)
ligand = mda.Universe(lig_prmtop, lig_inpcrd)
```

```python
for charge in receptor.atoms.charges:
    if charge < 0:
        print(charge)
```

```python
total_protein_charge = 0.0
print("Receptor Negative Residues")
for i in range(len(receptor.residues.resnames)):
    if receptor.residues.resnames[i] in ["ASP", "GLU"]:
        print(receptor.residues.resnames[i], i)
        total_protein_charge -= 1.
print("Ligand Positive Residues")
for i in range(len(ligand.residues.resnames)):
    if ligand.residues.resnames[i] in ["LYS", "ARG", "HIS"]:
        print(ligand.residues.resnames[i], i)
        total_protein_charge += 1.
```

```python
print("Receptor Positive Residues")
for i in range(len(receptor.residues.resnames)):
    if receptor.residues.resnames[i] in ["LYS", "ARG", "HIS"]:
        print(receptor.residues.resnames[i], i)
        total_protein_charge += 1.
print("Ligand Negative Residues")
for i in range(len(ligand.residues.resnames)):
    if ligand.residues.resnames[i] in ["ASP", "GLU"]:
        print(ligand.residues.resnames[i], i)
        total_protein_charge -= 1.
print(total_protein_charge)
```

```python
import numpy as np
def calculate_residue_charge(residue):
    charge = np.sum(residue.atoms.charges)
    return charge

total_charge = 0.0
for residue in receptor.residues:
    residue_charge = calculate_residue_charge(residue)
    total_charge += residue_charge
    print(residue.resname, residue_charge, total_charge)
```
